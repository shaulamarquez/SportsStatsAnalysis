```python
# drop the duplicates
athlete_events = athlete_events.drop_duplicates()
```

```python
# How many athletes participated in each Olympic event?
event_counts = athlete_events['Sport'].value_counts()
event_counts_df = event_counts.reset_index()
event_counts_df.columns = ['sport', 'athlete_count']
event_counts_df = event_counts_df.sort_values(by='athlete_count', ascending=False)
event_counts_df.head()
```

```python
# What are the average age, height, and weight of athletes by sport?
average_stats_by_sport = athlete_events.groupby('Sport').agg({
    'Age': 'mean',
    'Height': 'mean',
    'Weight': 'mean'
}).reset_index()

average_stats_by_sport.columns = ['Sport', 'Average_Age', 'Average_Height', 'Average_Weight']

average_stats_by_sport = average_stats_by_sport.dropna()

average_stats_by_sport
```

```python
# What is the gender distribution of athletes across different sports?
gender_distribution = athlete_events.groupby(['Sport', 'Sex']).size().reset_index(name='Count')
gender_distribution_pivot = gender_distribution.pivot(index='Sport', columns='Sex', values='Count').fillna(0)

gender_distribution
```

```python
# Which countries (NOCs) have the most participants in the Olympics?
participants_by_country = athlete_events['NOC'].value_counts().reset_index()
participants_by_country.columns = ['NOC', 'participants_by_country']
participants_by_country = participants_by_country.sort_values(by='participants_by_country', ascending=False)
participants_by_country
```

```python
# How many athletes participated in each year of the Olympics?
athletes_per_year = athlete_events.groupby('Year').size().reset_index(name='Number_of_Athletes')
athletes_per_year = athletes_per_year.sort_values(by='Number_of_Athletes', ascending=False)
athletes_per_year
```

```python
# How many medals by Country and Type

# Filter out rows where 'Medal' is NaN (no medal)
medals = athlete_events.dropna(subset=['Medal'])

# Count the number of each type of medal for each NOC
medals_by_noc_type = medals.groupby(['NOC', 'Medal']).size().unstack(fill_value=0).reset_index()

# Rename columns for clarity
medals_by_noc_type.columns.name = None
medals_by_noc_type.columns = ['NOC', 'Gold', 'Silver', 'Bronze']

medals_by_noc_type = medals_by_noc_type.sort_values(by=['Gold', 'Silver', 'Bronze'], ascending=False)
medals_by_noc_type
```
